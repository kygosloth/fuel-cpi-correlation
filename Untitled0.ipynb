```python
import pandas as pd

# Load the CSV into a DataFrame
df = pd.read_csv('fuel_prices_nairobi.csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Display the first few rows to verify
print("First 5 rows of the dataset:")
print(df.head())

# Check data types and missing values
print("\nData Info:")
print(df.info())
```

```python
import matplotlib.pyplot as plt

# Load your DataFrame (assuming it's already loaded as 'df' from fuel_prices_nairobi.csv)
# df = pd.read_csv('fuel_prices_nairobi.csv')
# df['Date'] = pd.to_datetime(df['Date'])

# Set figure size
plt.figure(figsize=(12, 6))

# Plot fuel prices on the left y-axis
ax1 = plt.gca()  # Get current axis
ax1.plot(df['Date'], df['Super_Petrol'], label='Super Petrol (KSh)', color='blue')
ax1.plot(df['Date'], df['Diesel'], label='Diesel (KSh)', color='green')
ax1.plot(df['Date'], df['Kerosene'], label='Kerosene (KSh)', color='red')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price (KSh)', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim(140, 210)  # Set appropriate range for fuel prices (based on data: ~150–200 KSh)

# Plot CPI metrics on the right y-axis
ax2 = ax1.twinx()  # Create a second y-axis that shares x-axis with current axis
ax2.plot(df['Date'], df['CPI'], label='CPI (Nairobi)', color='purple', linestyle='--')
ax2.plot(df['Date'], df['Transport_CPI'], label='Transport CPI', color='orange', linestyle='--')
ax2.plot(df['Date'], df['Energy_CPI'], label='Energy CPI', color='brown', linestyle='--')
ax2.set_ylabel('CPI Index', color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.set_ylim(140, 150)  # Set appropriate range for CPI (based on data: ~142–145.5)

# Title and legend
plt.title('Fuel Prices vs. CPI in Nairobi (Feb 2024–Feb 2025)')
ax1.legend(loc='upper left', bbox_to_anchor=(0, 0.9))
ax2.legend(loc='upper left', bbox_to_anchor=(0, 0.8))

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Add grid for clarity
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the plot with a clear filename
plt.savefig('fuel_prices_vs_cpi_corrected.png')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[['Super_Petrol', 'Diesel', 'Kerosene', 'CPI', 'Transport_CPI', 'Energy_CPI']].corr()

print("Correlation Matrix:")
print(correlation_matrix)

# Focus on correlations with CPI
print("\nCorrelations with CPI:")
print(correlation_matrix['CPI'][['Super_Petrol', 'Diesel', 'Kerosene']])
```

```python
# Correlation matrix
correlation_matrix = df[['Super_Petrol', 'Diesel', 'Kerosene', 'CPI', 'Transport_CPI', 'Energy_CPI']].corr()

print("Correlation Matrix:")
print(correlation_matrix)

# Correlations with CPI
print("\nCorrelations with CPI:")
print(correlation_matrix['CPI'][['Super_Petrol', 'Diesel', 'Kerosene']])

# Correlations with fuel prices (Exchange Rate and Murban Crude)
print("\nCorrelations with Fuel Prices:")
print(df[['Super_Petrol', 'Diesel', 'Kerosene', 'Exchange_Rate', 'Murban_Crude']].corr())
```

```python
from sklearn.linear_model import LinearRegression

# Define independent (X) and dependent (y) variables
X = df[['Super_Petrol', 'Diesel', 'Kerosene']]
y = df['CPI']

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Print coefficients, intercept, and R-squared
print("Regression Coefficients:")
for fuel, coef in zip(['Super_Petrol', 'Diesel', 'Kerosene'], model.coef_):
    print(f"{fuel}: {coef:.4f}")
print(f"Intercept: {model.intercept_:.4f}")

# Predict CPI change for a KSh 5 increase in each fuel
increase = 5
pred_change = sum(model.coef_ * increase)
print(f"\nA KSh 5 increase in all fuels raises CPI by {pred_change:.2f} points")

# R-squared to check model fit
r_squared = model.score(X, y)
print(f"R-squared: {r_squared:.4f}")
```

```python
# Plot Exchange Rate and Murban Crude
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Exchange_Rate'], label='Exchange Rate (KSh/USD)', color='blue')
plt.plot(df['Date'], df['Murban_Crude'], label='Murban Crude (USD/barrel)', color='red')
plt.title('Exchange Rate vs. Murban Crude (Feb 2024–Feb 2025)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.savefig('exchange_rate_vs_crude.png')
plt.show()
```
